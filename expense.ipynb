```python
from faker import Faker
import random
import pandas as pd
```

```python
fake=Faker()
```

```python
Category=["groceries","stationary","bills","subscriptions","investments",
          "transportation","school fees","home essential","sports&fitness",
          "fruits","vegetables","medical","emergency funds","parking","travel",
          "gas","care products","memberships","clothing"]
Payment_mode=["upi","debit card","cash","wallet","prepaid cards","net banking",
              "mobile payments","digital wallets","credit card"]
```

```python
from datetime import datetime, timedelta
def expdata(num_entries=110, year=None, month=None):
    if year is None:
        year = datetime.now().year
    if month is None:
        month = datetime.now().month
    data = []
    start_date = datetime(year, month, 1)
    end_date = (start_date + timedelta(days=32)).replace(day=1) - timedelta(days=1)  
    for _ in range(num_entries):  
        expense_tracker = {
            "Date": fake.date_between_dates(start_date, end_date),  
            "Category": random.choice(Category),
            "Payment_mode": random.choice(Payment_mode),
            "Description": fake.sentence(),
            "Amount": round(random.uniform(10, 100), 2),
            "Cashback": round(random.uniform(1, 10), 2),  
        }
        data.append(expense_tracker)
    return pd.DataFrame(data)
monthly_data = {}
for month in range(1, 13):  
    monthly_data[f"Month_{month}"] = expdata(num_entries=110, year=datetime.now().year, month=month)
monthly_data["Month_1"]
```

```python
combined_data = pd.concat(monthly_data.values(), keys=monthly_data.keys())
combined_data
```

```python
import csv
```

```python
df=pd.DataFrame(combined_data)
output_file="output.csv"
df.to_csv(output_file,index=False)
print(f"DataFrame successfully written to{output_file}")
```
